```python
import requests
import json
from adapters.remoteOK_adapter import RemoteOKAdapter
from database import Database
```

```python
url = 'https://remoteok.com/api'

response = requests.get(url)

# print(response.json())
jobs = response.json()
```

```python
with open("jobs.json", "w") as f:
    json.dump(jobs[1:], f, indent=4)
```
