```python
import requests

# Get latest Who is Hiring thread
r = requests.get(
    "https://hn.algolia.com/api/v1/search?query=Ask HN: Who is hiring?&tags=story,author_whoishiring"
)
print(r.json()['hits'][0]['title'])  # Should show latest thread

# Get thread ID
thread_id = r.json()['hits'][0]['objectID']

# Get all jobs
jobs = requests.get(f"https://hn.algolia.com/api/v1/items/{thread_id}").json()
print(f"Found {len(jobs['children'])} jobs")
```
